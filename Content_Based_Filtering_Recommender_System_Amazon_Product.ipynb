```
import pandas as pd
```

# Import Data

```
#Movies information
Amazon_df = pd.read_csv('/content/drive/My Drive/MarkovApplied/Section5/Amazon_products_reviews.csv')
Amazon_df.head()
```

```
AmazonFilter_df = Amazon_df.drop(['reviews.sourceURLs', 'reviews.text','reviews.username','reviews.userProvince',
                                  'reviews.userCity','reviews.title','reviews.rating','reviews.numHelpful',
                                  'reviews.id','reviews.doRecommend','reviews.didPurchase','reviews.dateSeen',
                                  'reviews.dateAdded','reviews.date','manufacturer','keys'], axis=1)
AmazonFilter_df.head()
```

```
AmazonFilter_df['categories'] = AmazonFilter_df['categories'].str.split(',')
AmazonFilter_df
```

```
AmazonV1_df =AmazonFilter_df.copy().dropna()
AmazonV12_df = AmazonV1_df.copy()
AmazonV12_df
```

```
for index, row in AmazonV12_df.iterrows():
  for categorie in row['categories']:
    AmazonV12_df.at[index,categorie] = 1 
AmazonV12_df.head()
```

```
AmazonV2_df = AmazonV12_df.fillna(0)
AmazonV2Category_df = AmazonV2_df.copy()
AmazonV2_df = (AmazonV2_df
           .drop_duplicates(subset=['name'], keep='last'))
AmazonV2_df.head()
```

```
AmazonV2Reviews_df = (AmazonV1_df
           .drop_duplicates(subset=['name'], keep='last'))
AmazonV2Reviews_df
```

```
AmazonV2Reviews_df['name'].value_counts()
```

# Get input user ratings

```
AmazonV2Reviews_df['name'].unique()
```

```
userInput = [
             {'name':'All-New Fire HD 8 Tablet, 8 HD Display, Wi-Fi, 16 GB - Includes Special Offers, Magenta','rating':3},
             {'name':'Amazon Kindle Fire Hd (3rd Generation) 8gb,,,\r\nAmazon Kindle Fire Hd (3rd Generation) 8gb,,,','rating':2},
             {'name':'Kindle Dx Leather Cover, Black (fits 9.7 Display, Latest and 2nd Generation Kindle Dxs),,','rating':4},
             {'name':'Fire HD 8 Tablet with Alexa, 8 HD Display, 16 GB, Tangerine - with Special Offers,','rating':5.5}             
    ]
inputProduct = pd.DataFrame(userInput)
inputProduct
```

```
inputId = AmazonV2Reviews_df[AmazonV2Reviews_df['name'].isin(inputProduct['name'].tolist())]
inputId
```

```
inputProductV2 = pd.merge(inputId, inputProduct)

inputProductV2
```

```
inputProductV3= inputProductV2.drop(['categories', 'brand','asins'], axis=1)
inputProductV3
```

# Get input user profile

```
userinputProduct = AmazonV2_df[AmazonV2_df['name'].isin(inputProductV3['name'].tolist())]
userinputProduct
```

```
userinputProduct = userinputProduct.reset_index(drop=True)
userCategoryTable = userinputProduct.drop(['id', 'name', 'categories', 'brand','asins'], axis=1)
userCategoryTable
```

```
inputProductV3['rating']
userCategoryTable.shape
```

```
userProfile = userCategoryTable.transpose().dot(inputProductV3['rating'])
#The user profile
userProfile
```

#  Find the recommendations

```
AmazonV2Recommendation_df =AmazonV2_df.copy()
AmazonV2Recommendation_df
```

```
#Now let's get the genres of every movie in our original dataframe
categoryTable = AmazonV2Recommendation_df.set_index(AmazonV2Recommendation_df['name'])
#And drop the unnecessary columns
categoryTable = categoryTable.drop(['id', 'name', 'categories','asins','brand'], axis=1)
categoryTable
```

```
((categoryTable*userProfile).sum(axis=1))/(userProfile.sum())
```

```
recommendationTable_df = ((categoryTable*userProfile).sum(axis=1))/(userProfile.sum())
recommendationTable_df
```

```
recommendationTableWeighted_df = (recommendationTable_df.reset_index()).rename(columns={0: 'weightedAvg'})
recommendationTableWeighted_df
```

```
user_recommendationTable_df = pd.merge(AmazonV2Reviews_df, recommendationTableWeighted_df)
user_recommendationTable_df
```

```
#Sort our recommendations in descending order
user_recommendationTable_df = user_recommendationTable_df.sort_values(by='weightedAvg', ascending=False)
user_recommendationTable_df.head(20)
```
